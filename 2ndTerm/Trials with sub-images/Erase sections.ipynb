```python
from ipynb.fs.full.Import_data  import import_data
from ipynb.fs.full.workflow     import workflow
```

```python
PCA_COMPONENTS = 20
LEARNING_RATE  = 0.01
EPOCHS         = 5000
SUBIMG_SPACING = 10
IMG_WIDTH      = 20
DROPOUT_RATE   = 0.1
TRAIN_CUTOFF   = 3750
```

```python
tissue = import_data()
```

## Use small sections from initial spectra

```python
sections = range(0,1024,64)

for start in sections:
    
    this_tissue = []
    for (t_map,t_target,width,height) in tissue: 
        t_map = t_map[:,start:start+64]
        this_tissue.append((t_map,t_target,width,height))
        
    workflow(this_tissue, PCA_COMPONENTS, IMG_WIDTH, SUBIMG_SPACING, DROPOUT_RATE, LEARNING_RATE, EPOCHS, TRAIN_CUTOFF)
```
